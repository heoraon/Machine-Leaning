```python
import urllib.request
import json
import csv
import datetime
import time

client_id = "Rwc1QbXu6J3mZzO8cXlG"
client_secret = "GxoJtGjiV7"

keyword = "AI ë””ì§€í„¸ êµê³¼ì„œ"
display = 100
max_start = 1000
start_date = datetime.date(2023, 1, 1)
end_date = datetime.date.today()


def parse_pubdate(pubdate_str):
    return datetime.datetime.strptime(pubdate_str, "%a, %d %b %Y %H:%M:%S +0900").date()

all_news = []
current_date = start_date

print("ì „ì²´ ë‰´ìŠ¤ ìˆ˜ì§‘ ì‹œì‘...")

while current_date <= end_date:
    print(f"{current_date} ìˆ˜ì§‘ ì¤‘...")
    daily_news = []
    encText = urllib.parse.quote(keyword)
    start = 1

    while start < max_start:
        url = f"https://openapi.naver.com/v1/search/news?query={encText}&display={display}&start={start}&sort=date"
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)

        try:
            response = urllib.request.urlopen(request)
            if response.getcode() != 200:
                print(f"ì‘ë‹µ ì½”ë“œ: {response.getcode()} - ì¤‘ë‹¨")
                break

            data = json.loads(response.read().decode('utf-8'))
            items = data.get('items', [])

            if not items:
                break

            for item in items:
                pubdate = parse_pubdate(item['pubDate'])
                if pubdate == current_date:
                    daily_news.append({
                        'title': item['title'],
                        'link': item['link'],
                        'description': item['description'],
                        'pubDate': item['pubDate']
                    })
                elif pubdate < current_date:
                    start = max_start  # ì´ì „ ë‚ ì§œê°€ ë‚˜ì˜¤ë©´ ì¤‘ì§€
                    break

            start += display
            time.sleep(0.3)  # API í˜¸ì¶œ ì œí•œ íšŒí”¼ìš© ë”œë ˆì´
        except Exception as e:
            print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")
            break

    print(f"{current_date} ìˆ˜ì§‘ëœ ë‰´ìŠ¤ ê°œìˆ˜: {len(daily_news)}")
    all_news.extend(daily_news)
    current_date += datetime.timedelta(days=1)

filename = f"naver_ai_textbook_news_daily_{start_date}_{end_date}.csv"

with open(filename, 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.DictWriter(f, fieldnames=['title', 'link', 'description', 'pubDate'])
    writer.writeheader()
    writer.writerows(all_news)

print(f"\nì´ {len(all_news)}ê°œì˜ ë‰´ìŠ¤ ìˆ˜ì§‘ ì™„ë£Œ. íŒŒì¼ëª…: {filename}")
```
