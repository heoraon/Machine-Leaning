```python
import math #파이썬 사용자 정의 함수(sigmoid 함수 정의)위해 사용
import random #학습에 영향을 줄 수 이쓴 a ~ i까지의 초기값 랜덤으로 부여하기 위해 사용
```

```python
#파라미터 값 랜덤 부여(random의 uniform함수: 괄호 안의 값 범위에서 랜덤한 값 리턴하는 함수)

#은닉층1 파라미터 값 초기화
a1 = random.uniform(-1, 1)
b1 = random.uniform(-1, 1)
c1 = random.uniform(-1, 1)
a2 = random.uniform(-1, 1)
b2 = random.uniform(-1, 1)
c2 = random.uniform(-1, 1)
a3 = random.uniform(-1, 1)
b3 = random.uniform(-1, 1)
c3 = random.uniform(-1, 1)

#은닉층2 파라미터 값 초기화
d1 = random.uniform(-1, 1)
e1 = random.uniform(-1, 1)
f1 = random.uniform(-1, 1)
d2 = random.uniform(-1, 1)
e2 = random.uniform(-1, 1)
f2 = random.uniform(-1, 1)

#은닉층3 파라미터 값 초기화
g1 = random.uniform(-1, 1)
h1 = random.uniform(-1, 1)
i1 = random.uniform(-1, 1)
g2 = random.uniform(-1, 1)
h2 = random.uniform(-1, 1)
i2 = random.uniform(-1, 1)

#출력층 파라미터 값 초기화
j  = random.uniform(-1, 1)
k  = random.uniform(-1, 1)
l  = random.uniform(-1, 1)
```

```python
eta = 0.0001 #파이썬 예약어 - 학습률을 지정(경사하강법에서 한 번에 이동하는 값)
```

```python
#sigmoid 함수(활성화 함수) 정의
'''
한 뉴런에서 임계치 이상의 신호가 올 때
다음 뉴런에 신호를 전달하는 기능을 구현하기 위해
특정값 이상에서 값이 크게 증가하는 시그모이드 함수를 활용
'''
def sigmoid(x):
  return 1.0/(1.0+math.exp(-x))
```

```python
#데이터셋 불러오기
from sklearn.datasets import load_breast_cancer
import numpy as np

dataset = load_breast_cancer()
```

```python
#데이터셋 형태 확인
print(dataset.data[:5])       # 첫 5개의 데이터
print(dataset.target[:5])     # 첫 5개의 타겟(레이블)
print(dataset.DESCR)          # 데이터셋 설명
```

```python
#데이터셋 로드
X_all = dataset.data
y_all = dataset.target  #0: 악성(malignant), 1: 양성(benign)

#사용할 feature 선택 (x, y로 사용할 두 feature)
x_idx = list(dataset.feature_names).index('mean radius')
y_idx = list(dataset.feature_names).index('mean texture')

#(x, y) 데이터 추출
data = [(X_all[i][x_idx], X_all[i][y_idx]) for i in range(len(X_all))]

#레이블 변환 (0 -> -1, 1 -> +1)
labels = [1 if y == 1 else -1 for y in y_all]
```

```python
#예측 결과 저장을 위한 초기화 리스트 생성
zValue = [0] * len(data)
FinalzValue = [0] * len(data)
```

```python
#모델 구현 및 학습
for counter in range(100000):
    #가중치/편향 기울기 초기화
    da1=db1=dc1=da2=db2=dc2=da3=db3=dc3=0.0 #입력층
    dd1=de1=df1=dd2=de2=df2=0.0 #은닉층1
    dg1=dh1=di1=dg2=dh2=di2=0.0 #은닉층2
    dj=dk=dl=0.0 #출력층

    for idx in range(len(data)): #데이터셋 길이만큼 학습 반복
        x=float(data[idx][0]) #데이터(두 쌍의 데이터 중 첫째 값)
        y=float(data[idx][1]) #데이터(두 쌍의 데이터 중 둘째 값)
        Z = labels[idx] #레이블 값

        #은닉층1 (3 뉴런)
        z1_1 = a1*x + b1*y + c1
        X1 = sigmoid(z1_1)
        z1_2 = a2*x + b2*y + c2
        X2 = sigmoid(z1_2)
        z1_3 = a3*x + b3*y + c3
        X3 = sigmoid(z1_3)

        #은닉층2 (2 뉴런)
        z2_1 = d1*X1 + e1*X2 + f1*X3 + f1  #편향 f1
        Y1 = sigmoid(z2_1)
        z2_2 = d2*X1 + e2*X2 + f2*X3 + f2  #편향 f2
        Y2 = sigmoid(z2_2)

        #은닉층3 (2 뉴런)
        z3_1 = g1*Y1 + h1*Y2 + i1
        Z1 = sigmoid(z3_1)
        z3_2 = g2*Y1 + h2*Y2 + i2
        Z2 = sigmoid(z3_2)

        #출력층 (1 뉴런)
        z = j*Z1 + k*Z2 + l

#역전파(백프로퍼게이션)
        #출력 뉴런에서의 오차 (MSE: Mean Squared Error의 미분)
        dz4 = 2*(z - Z)

        #출력층 가중치 기울기
        dj += dz4 * Z1
        dk += dz4 * Z2
        dl += dz4

        #은닉층3 기울기
        dz3_1 = dz4 * j * sigmoid(z3_1)*(1 - sigmoid(z3_1))
        dz3_2 = dz4 * k * sigmoid(z3_2)*(1 - sigmoid(z3_2))

        dg1 += dz3_1 * Y1
        dh1 += dz3_1 * Y2
        di1 += dz3_1

        dg2 += dz3_2 * Y1
        dh2 += dz3_2 * Y2
        di2 += dz3_2

        #은닉층2 기울기
        dz2_1 = (dz3_1 * g1 + dz3_2 * g2) * sigmoid(z2_1)*(1 - sigmoid(z2_1))
        dz2_2 = (dz3_1 * h1 + dz3_2 * h2) * sigmoid(z2_2)*(1 - sigmoid(z2_2))

        dd1 += dz2_1 * X1
        de1 += dz2_1 * X2
        df1 += dz2_1 * X3
        dd2 += dz2_2 * X1
        de2 += dz2_2 * X2
        df2 += dz2_2 * X3

        #은닉층1 기울기
        dz1_1 = (dz2_1 * d1 + dz2_2 * d2) * sigmoid(z1_1)*(1 - sigmoid(z1_1))
        dz1_2 = (dz2_1 * e1 + dz2_2 * e2) * sigmoid(z1_2)*(1 - sigmoid(z1_2))
        dz1_3 = (dz2_1 * f1 + dz2_2 * f2) * sigmoid(z1_3)*(1 - sigmoid(z1_3))

        da1 += dz1_1 * x
        db1 += dz1_1 * y
        dc1 += dz1_1

        da2 += dz1_2 * x
        db2 += dz1_2 * y
        dc2 += dz1_2

        da3 += dz1_3 * x
        db3 += dz1_3 * y
        dc3 += dz1_3

    #기울기 누적합을 활용해 파라미터 업데이트
    a1 -= eta * da1
    b1 -= eta * db1
    c1 -= eta * dc1
    a2 -= eta * da2
    b2 -= eta * db2
    c2 -= eta * dc2
    a3 -= eta * da3
    b3 -= eta * db3
    c3 -= eta * dc3

    d1 -= eta * dd1
    e1 -= eta * de1
    f1 -= eta * df1
    d2 -= eta * dd2
    e2 -= eta * de2
    f2 -= eta * df2

    g1 -= eta * dg1
    h1 -= eta * dh1
    i1 -= eta * di1
    g2 -= eta * dg2
    h2 -= eta * dh2
    i2 -= eta * di2

    j -= eta * dj
    k -= eta * dk
    l -= eta * dl

#최종적으로 업데이트 된 파라미터 값 출력
print("*** 은닉층1 가중치 및 편향 ***")
print(f"a1={a1:.6f}, b1={b1:.6f}, c1={c1:.6f}")
print(f"a2={a2:.6f}, b2={b2:.6f}, c2={c2:.6f}")
print(f"a3={a3:.6f}, b3={b3:.6f}, c3={c3:.6f}\n")

print("*** 은닉층2 가중치 및 편향 ***")
print(f"d1={d1:.6f}, e1={e1:.6f}, f1={f1:.6f}")
print(f"d2={d2:.6f}, e2={e2:.6f}, f2={f2:.6f}\n")

print("*** 은닉층3 가중치 및 편향 ***")
print(f"g1={g1:.6f}, h1={h1:.6f}, i1={i1:.6f}")
print(f"g2={g2:.6f}, h2={h2:.6f}, i2={i2:.6f}\n")

print("*** 출력층 가중치 및 편향 ***")
print(f"j={j:.6f}, k={k:.6f}, l={l:.6f}")
```

```python
#예측값 누적 배열 초기화 (데이터 수 만큼 0으로 초기화)
zValue = [0.0 for _ in range(len(data))]
FinalzValue = [0.0 for _ in range(len(data))]

for counter in range(100000):  #100만 번 반복 (예측값 안정화 목적)
    for idx in range(len(data)):
        x = float(data[idx][0])  #입력값 x
        y = float(data[idx][1])  #입력값 y

        #은닉층1 뉴런 3개
        z1_1 = a1*x + b1*y + c1
        X1 = sigmoid(z1_1)

        z1_2 = a2*x + b2*y + c2
        X2 = sigmoid(z1_2)

        z1_3 = a3*x + b3*y + c3
        X3 = sigmoid(z1_3)

        #은닉층2 뉴런 2개
        z2_1 = d1*X1 + e1*X2 + f1*X3 + f1
        Y1 = sigmoid(z2_1)

        z2_2 = d2*X1 + e2*X2 + f2*X3 + f2
        Y2 = sigmoid(z2_2)

        #은닉층3 뉴런 2개
        z3_1 = g1*Y1 + h1*Y2 + i1
        Z1 = sigmoid(z3_1)

        z3_2 = g2*Y1 + h2*Y2 + i2
        Z2 = sigmoid(z3_2)

        #출력층 뉴런 1개 (선형 활성 함수 가정)
        z4 = j*Z1 + k*Z2 + l
        output = z4  # 최종 예측값

        #예측값 누적
        zValue[idx] += output

#반복 종료 후 평균 예측값 계산
for idx in range(len(data)):
    FinalzValue[idx] = zValue[idx] / 1000000

#최종 예측값 출력
print(FinalzValue)
```
